```python
from ultralytics import YOLO
import pandas as pd
import onnxruntime as ort
import cv2
import numpy as np
from matplotlib import pyplot as plt
from sklearn import metrics

from pathbook.pathbook import *
```

```python
results = YOLO('../bin/yolov8x-seg.onnx').predict(path_test_dataset,imgsz=320,batch=10)
df = pd.read_csv(path_test_annotation)
```

```python
y_pred = []
y_true = []

for result in results:
    bboxes = pd.DataFrame(result.boxes.data, columns=['x1','y1','x2','y2','conf','class'])
    y_pred.append(result.names[bboxes.iloc[bboxes['conf'].idxmax()]['class']])

    answer_df = df[df.path == result.path.split('/')[-1]]['label']
    y_true.append(answer_df.iloc[answer_df.count().argmax()])

print(metrics.accuracy_score(y_true, y_pred))
```
